#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# code
df = pd.DataFrame(data)
x_values = df.loc[:,"age"]#truy van nhan age
y_values = df.loc[:,"refletter.rating"]
plt.xlabel('age')#dat nhan age cho truc x
plt.ylabel('refletter.rating')
plt.scatter(x_values, y_values, s=10, color="blue") #ve bieu do
plt.title('relationship between age and refletter.rating')
plt.show()
```

#7. Use scatter plot to represent the relationship between ***age*** and ***sciences.grade***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# code
df = pd.DataFrame(data)
x_values = df.loc[:,"age"]
y_values = df.loc[:,"sciences.grade"]
plt.xlabel('age')
plt.ylabel('sciences.grade')
plt.scatter(x_values, y_values, s=10, color="blue")
plt.title('Relationship between age and sciences.grade')
plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***
using bins=5

```python
#code
ages=df.loc[:,"age"]
range = (18, 70)
bins = 5
plt.title('Distribution of Age')
plt.hist(ages, bins, range, color='blue',histtype='bar',rwidth=0.5)
plt.xlabel('age')
plt.ylabel('')
plt.show()
```

#9. Create a data frame that computes the average of ages in each countries?
